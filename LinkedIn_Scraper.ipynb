```python

for i in range(len(url_list)):
    browser.get(url_list[i])
    browser.implicitly_wait(3)
    
    webpage=requests.get(url_list[i]).text
    
    soup_list.append(BeautifulSoup(webpage,'lxml'))
    #print(soup.prettify())
    
    
    
    for link in soup_list[i].find_all("a"):
        href_link_list.append(link.get("href"))
        inner_text.append("Inner Text: {}".format(link.text.strip()))
       
```

```python

for i in range(3439,len(href_link_list)):
    if 'jobs/view' in str(href_link_list[i]):
        href_link_final.append(href_link_list[i])
    else:
        continue
```

```python

count=0
for link in range(1350,len(href_link_final)):
    count+=1
    print(count)
    browser.get(href_link_final[link])
    browser.implicitly_wait(3)
    
    try:
        job_title = browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__job-title")
        Job_title.append(job_title[0].text)
    except:
        Job_title.append('NA')
        
    try:    
        company_name=browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__company-name")
        Company_name.append(company_name[0].text)
    except:
        Company_name.append('NA')
        
    try:    
        location_name=browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__bullet")
        Location_name.append(location_name[0].text)
    except:
        Location_name.append('NA')
        
    try:
        time_posted=browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__posted-date")
        Time_posted.append(time_posted[0].text)
    except:
        Time_posted.append('NA')
    
    try:
        applicants_count=browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__applicant-count")
        Applicants_count.append(applicants_count[0].text)
    except:
        Applicants_count.append('NULL')

    try:
        job_nature_level=browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__job-insight")
        Job_nature_level.append(job_nature_level[0].text)
        Emp_count_Ctype.append(job_nature_level[1].text)
    except:
        Job_nature_level.append('NA')
        Emp_count_Ctype.append('NA')
    
    try:
        emp_count_onLinkedin=browser.find_elements(By.CLASS_NAME,'jobs-company__inline-information')
        Emp_count.append(emp_count_onLinkedin[0].text)
        Emp_on_Linkedin.append(emp_count_onLinkedin[1].text)
    except:
        Emp_count.append('NA')
        Emp_on_Linkedin.append('NA')
    
    try:
        pay_range= browser.find_elements(By.CLASS_NAME,'jobs-unified-description__salary-main-rail-card')
        Pay_range.append(pay_range[0].text.strip())
    except:
        Pay_range.append('NA')
        
    try:    
        followers=browser.find_elements(By.CLASS_NAME,'artdeco-entity-lockup__subtitle')
        Company_followers.append(followers[0].text)
    except:
        Company_followers.append('NA')
```
